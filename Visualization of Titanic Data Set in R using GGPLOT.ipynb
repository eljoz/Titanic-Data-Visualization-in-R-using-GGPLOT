### Data Importation

```r
#Loading necessary libraries
library(tidyverse)
```

```r
data = read.csv("train.csv")

options(repr.plot.width = 15, repr.plot.height=5)
```

```r
df<-as.data.frame(data)
```

```r
head(df)
```

```r
tail(df)
```

```r
dim(df)
```

```r
str(df)
```

```r
summary(df)
```

```r
df2 <- df %>%
    select(-1) %>%
    mutate(Survived = factor(Survived, levels = c(0,1), labels = c("Died", "Survived")),
          Sex = factor(Sex),
          Pclass = factor(Pclass))
```

```r
head(df2)
```

### Data Visualization

```r
#Table
table(df2$Survived)
```

```r
#Graph
df2 %>%
  ggplot(aes(y = Survived)) +
  geom_bar( width = 0.8, fill="darkblue",color="red", position= "dodge", alpha= 1) + 
      scale_x_continuous(expand = c(0, NA))+
      theme(
          plot.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          plot.caption = element_text(family = "Times New Roman", hjust = 0.5, size = 30),
          axis.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          axis.text  = element_text(family = "Times New Roman", face = "bold", size = 15))+
      labs(title = "Overall Survival Count",
           caption = "source: http://titanic.com",
           x = "Passenger Count",
       
    )
```

```r
df2 %>%
  ggplot(aes(y = Survived, fill=Sex)) +
  geom_bar( width = 0.8,color="red", position= "dodge", alpha= 1) + 
      scale_x_continuous(expand = c(0, NA))+
      theme(
          plot.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          plot.caption = element_text(family = "Times New Roman", hjust = 0.5, size = 30),
          axis.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          axis.text  = element_text(family = "Times New Roman", face = "bold", size = 15))+
      labs(title = "Overall Survival Count By Gender",
           caption = "source: http://titanic.com",
           x = "Passenger Count",
       
    )
```

```r
df2 %>%
  ggplot(aes(y = Survived, fill=Pclass)) +
  geom_bar( width = 0.8,color="red", position= "dodge", alpha= 1) + 
      scale_x_continuous(expand = c(0, NA))+
      theme(
          plot.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          plot.caption = element_text(family = "Times New Roman", hjust = 0.5, size = 30),
          axis.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          axis.text  = element_text(family = "Times New Roman", face = "bold", size = 15))+
      labs(title = "Overall Survival Count By Passenger Class",
           caption = "source: http://titanic.com",
           x = "Passenger Count",
       
    )
```

```r
df2 %>%
  ggplot(aes(x = Age, fill=Survived)) +
  geom_histogram(bins=50, position= "dodge", alpha= 1) + 
      scale_x_continuous(expand = c(0, NA))+
      theme(
          plot.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          plot.caption = element_text(family = "Times New Roman", hjust = 0.5, size = 30),
          axis.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          axis.text  = element_text(family = "Times New Roman", face = "bold", size = 15))+
      labs(title = "Overall Survival Count By Passenger Age",
           caption = "source: http://titanic.com",
           y = "Passenger Count",
           x = NULL
       
    )
```

```r
df2 %>%
  ggplot(aes(y = Survived, x=Age)) +
  geom_boxplot( width = 0.8,color="red", position= "dodge", alpha= 1) + 
      scale_x_continuous(expand = c(0, NA))+
      theme(
          plot.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          plot.caption = element_text(family = "Times New Roman", hjust = 0.5, size = 30),
          axis.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          axis.text  = element_text(family = "Times New Roman", face = "bold", size = 15))+
      labs(title = "Overall Survival Count By Passenger Age",
           caption = "source: http://titanic.com",
           x = "Passenger Count",
       
    )
```

```r
df2 %>%
  ggplot(aes(y = Survived, fill=Sex)) +
  geom_bar( width = 0.8,color="red", position= "dodge", alpha= 1) + 
      facet_wrap(~ Pclass)+
      scale_x_continuous(expand = c(0, NA))+
      theme(
          plot.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          plot.caption = element_text(family = "Times New Roman", hjust = 0.5, size = 30),
          axis.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          axis.text  = element_text(family = "Times New Roman", face = "bold", size = 15))+
      labs(title = "Overall Survival Count By Passenger Class and Sex",
           caption = "source: http://titanic.com",
           x = "Passenger Count",
       
    )
```

```r
df2 %>%
  ggplot(aes(x = Age, fill=Survived)) +
  geom_bar( width = 0.8, position= "dodge", alpha= 1) + 
      facet_wrap(~ Pclass+Sex)+
      scale_x_continuous(expand = c(0, NA))+
      theme(
          plot.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          plot.caption = element_text(family = "Times New Roman", hjust = 0.5, size = 30),
          axis.title = element_text(family = "Times New Roman", face = "bold", size = 30),
          axis.text  = element_text(family = "Times New Roman", face = "bold", size = 15))+
      labs(title = "Overall Survival Count By Passenger Class and Sex and Age",
           caption = "source: http://titanic.com",
           y = "Passenger Count",
       
    )
```

more females withing the age group of 20 to 40 and in first class Survived

Also more males within the age group of 20 to 40 and in the third class Died
